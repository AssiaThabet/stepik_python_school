```python
import math

xxs = range(0,60,5)
xs = []
ys = []
for i in xxs:
    x = i/10 #создание x
    xs.append(x)
    y = math.sin(x) # вычислеие y 
    ys.append(y)
plt.plot(xs,ys)
```

```python
xxs = range(0,1000,5)
xs = []
ys = []
for i in xxs:
    x = i/100 #создание x
    xs.append(x)
    y = x*math.sin(x) # вычислеие y 
    ys.append(y)
plt.plot(xs,ys)
```

```python
f = open('coordinates.txt')
data = f.read()
lines = data.splitlines()
x = lines[0][1:].split()
y = lines[1][1:].split()
```

```python
xs = []
ys = []

for i in range(len(x)):
    xs.append(int(x[i]))

for i in range(len(y)):
    ys.append(int(y[i]))
```

```python
plt.plot(xs,ys)
```
